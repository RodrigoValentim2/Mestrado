```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("contagem_01.csv", sep="\t")
```

```python
df.head()
```

```python
import matplotlib.ticker as ticker
```

```python
palhetas = np.array(df['Nº do par'], dtype=str)
plt.figure(figsize=(8,5))
plt.plot(palhetas, df['Humano'],  color='blue', marker=1, label='Humano')
plt.plot(palhetas, df['Máquina'],  color='red', marker="o", label="Máquina");
plt.title("Grupo de palhetas 1")
plt.xlabel("Número do par", fontsize=15)
plt.ylabel("Número de ovos",fontsize=15 )
plt.ylim(0,max(df['Humano']), 10 )
plt.legend(loc='best');
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.plot(palhetas, df['Humano.1'],  color='blue', marker=1, label='Humano')
plt.plot(palhetas, df['Máquina.1'],  color='red', marker="o", label="Máquina");
plt.title("Grupo de palhetas 2")
plt.xlabel("Número do par", fontsize=15)
plt.ylabel("Número de ovos",fontsize=15 )
plt.ylim(0,max(df['Humano.1']), 10 )
plt.legend(loc='best');
```
